```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set width of bar
barWidth = 0.02
fig, axes = plt.subplots(1, 2, figsize =(12, 5))
 
# set height of bar
Train_acc = [0, 0.994232, 0.994232, 0]
Test_acc = [0, 0.416667, 0.488889, 0]
 
# Set position of bar on X axis
br1 = np.arange(len(Train_acc))
br1 = [0.175, 0.2, 0.3, 0.325]
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
axes[0].bar(br1, Train_acc, color ='b', alpha=0.7, width = barWidth,
        edgecolor ='black', linewidth=1, label ='Train')
axes[0].bar(br2, Test_acc, color ='r', alpha=0.6, width = barWidth,
        edgecolor ='black', linewidth=1, label ='Test')

# Adding configs
axes[0].set_ylabel('Worst-group accuracy',  fontsize = 25)
axes[0].set_xticks([r + barWidth * 0.5 for r in br1[1:3]])
#axes[0].tick_params(axis='both', which='major', labelsize=20)
axes[0].set_xticklabels(['standard', 'importance\nsampling'], fontsize=14)
axes[0].set_title('SGD', fontweight ='bold',  fontsize=22, y=1.01)
axes[0].set_ylim(0, 1.01)

# set height of bar
Train_acc = [0, 0.406205, 0.955347, 0]
Test_acc = [0, 0.294444, 0.850000, 0]

# Make the plot
axes[1].bar(br1, Train_acc, color ='b', alpha=0.7, width = barWidth,
        edgecolor ='black', linewidth=1, label ='Train')
axes[1].bar(br2, Test_acc, color ='r', alpha=0.6, width = barWidth,
        edgecolor ='black', linewidth=1, label ='Test')
#axes[1].set_ylabel('Worst-group accuracy', fontsize=25)
axes[1].set_xticks([r + barWidth * 0.5 for r in br1[1:3]])
#axes[1].tick_params(axis='both', which='major', labelsize=20)
axes[1].set_xticklabels(['standard', 'importance\nsampling'], fontsize=14)
axes[1].set_title('DP-SGD', fontweight ='bold',  fontsize=22, y=1.01)
axes[1].legend(fontsize=20, bbox_to_anchor=(0.5, 1.04), frameon=False)


#plt.rc('xtick', labelsize=14) 
#plt.rc('ytick', labelsize=14) 

plt.ylim(0.0, 1.01)
#plt.tick_params(axis='x', labelsize=23)
plt.tick_params(axis='y', labelsize=15)
axes[0].grid(axis='y', linestyle='--')
axes[1].grid(axis='y', linestyle='--')
plt.savefig('main_fig1.pdf', bbox_inches='tight')
```

```python
# set width of bar
barWidth = 0.02
fig = plt.subplots(figsize =(6, 7))
 
# set height of bar
Train_acc = [0, 0.406205, 0.955347, 0]
Test_acc = [0, 0.294444, 0.850000, 0]
 
# Set position of bar on X axis
br1 = np.arange(len(Train_acc))
br1 = [0.175, 0.2, 0.3, 0.325]
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, Train_acc, color ='b', alpha=0.7, width = barWidth,
        edgecolor ='black', linewidth=1, label ='Train')
plt.bar(br2, Test_acc, color ='r', alpha=0.6, width = barWidth,
        edgecolor ='black', linewidth=1, label ='Test')
 
# Adding configs
plt.ylabel('Accuracy',  fontsize = 25)
plt.xticks([r + barWidth * 0.5 for r in br1[1:3]], 
           ['without IS', 'with IS'],fontsize=20)
plt.title('With $\it{\,\,DP\,\, Training}$', fontweight ='bold',  fontsize = 22, y=1.01)
plt.rc('xtick', labelsize=14) 
plt.tick_params(axis='x', labelsize=23)
plt.tick_params(axis='y', labelsize=15)
plt.legend(fontsize=20, bbox_to_anchor=(1, 0.6))

plt.ylim(0.0, 1.01)
plt.grid(axis='y', linestyle='--')
plt.savefig('main_fig2.pdf', bbox_inches='tight')
```
