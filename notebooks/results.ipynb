# CelebA

```python
ds_name = "celebA"

algos = [
    "erm",
    "erm_reweight",
    "groupDRO",
    "groupDRO_wd0.1",
    #"erm-dp_resnet18-dpsgd_1e-5_1.0_1.0_0.00005",
    "weightederm-dp_resnet18-dpsgd_1e-5_1.0_1.0_0.00005",
    "weightederm-dp_resnet18-dpsgd_1e-5_0.1_1.0_0.00005",
    "subsamplederm-dp_resnet18-dpsgd_1e-5_1.0_1.0_0.005",
    #"groupdro-dp_resnet18-dpsgd_1e-5_0.1_1.0_0.00005"
    "iwerm-dp_resnet18-dpsgd_1e-5_1.0_0.1_0.0001",
    "iwerm-dp_resnet18-dpsgd_1e-5_1.0_1.0_0.0001",
    "iwerm-dp_resnet18-dpsgd_1e-5_1.0_10.0_0.0001",
    
]
cols = ["acc_avg", "acc_wg", "acc_y:notblond_male:0", "acc_y:notblond_male:1", "acc_y:blond_male:0", "acc_y:blond_male:1"]
algo_names = ["ERM",
              "ERM IW", "DRO", "DRO_wd", #"DP ERM",
              "DP IW", "DP IW2",
              "DP SUB",
              #"DP DRO",
              "DP IWERM", "DP IWERM1", "DP IWERM2",
             ]
col_names = ["acc", "acc_wg", "notblond_male:0", "notblond_male:1", "blond_male:0", "blond_male:1"]

trn_results = {}
tst_results = {}
for i, algo in enumerate(algos):
    print(algo)
    print(f"../logs/celebA/{algo}/test_eval.csv")
    df = pd.read_csv(f"../logs/{ds_name}/{algo}/test_eval.csv")
    tst_results[algo_names[i]] = df[cols].iloc[-1].tolist()
    
    df = pd.read_csv(f"../logs/{ds_name}/{algo}/train_eval.csv")
    trn_results[algo_names[i]] = df[cols].iloc[-1].tolist()
trn_df = pd.DataFrame.from_dict(trn_results, orient="index", columns=col_names)
tst_df = pd.DataFrame.from_dict(tst_results, orient="index", columns=col_names)
```
